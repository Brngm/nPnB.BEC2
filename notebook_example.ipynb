```python
# INPUTS ==================================================
nameGraph="GraphExample.txt"
sp=0.49
so=0.1
Epsilon=0.01
rankedEdeges="0"
InfileClustZero="0"
# ---------------------------------------------------------
InfileGraph = "./data/" + nameGraph
OutfileClust = "./data/CLUSTERING/" + nameGraph + ".clustOutput.txt" 
# =========================================================

EXIT, ERROR, X = nPnB.BEC(InfileGraph, OutfileClust=OutfileClust, sp=sp, so=so,
                               Epsilon=Epsilon, InfileClustZero=InfileClustZero, rankedEdeges=rankedEdeges, Verbose="1")
if (EXIT!=0):
    print(ERROR)
else:
    # -------------------------------------------------------------
    date=X["date"] # computing date
    commandline="bec "+X["bec"] # command line

    input_InfileGraph=X["input_InfileGraph"] # input file graph
    input_OutfileClust=X["input_OutfileClust"] # output file clustering

    input_RankedEdeges=X["input_RankedEdeges"] # input RankedEdeges used for computing the community structure
    input_ClustZero=X["input_ClustZero"] # input ClustZero used for computing the community structure
    input_epsilon=X["input_epsilon"] # input epsilon used for computing the community structure
    
    input_sp=X["input_sp"] # input description scale used to construct Cp the clustering by partition
    input_so=X["input_so"]  # input stickiness scale, overlap amount used for the extension of the partition modules

    gnbv_call=X["gnbv"] # number of nodes in the graph on which the community structure were built
    gnbe_call=X["gnbe"] # number of edges in the graph on which the community structure were built
    density_call=X["density"] # edges density in the graph on which the community structure were built
    
    secondsP=X["secondsP"] # Time computing for partition
    secondsO=X["secondsO"] # Time computing for the extensions
    
    Cp=X["Cp"] # the clustering by partition
    Ext=X["Ext"] # the gluant extentions of the modules of Cp: len(Cp) = len(Ext). For all i, Cp[i] ∩ Ext[i] = ∅
    Co=[Cp[i]+Ext[i] for i in range(len(Cp))] # the clustering alowing overlaps

    # intrinsic metrics of Cp with respect to the graph on which it was constructed
    # A node i is adherent to a module m if i ∉ m and i is a neighbor of all the nodes of m
    omega_p=X["omega_p"] # // sum(|adherent(m)|) for m in CP
    TP_p=X["TP_p"] # True positive of CP
    TN_p=X["TN_p"] # True negative  of CP
    FP_p=X["FP_p"] # False positive  of CP
    FN_p=X["FN_p"] # False negative  of CP
    P_p=X["P_p"] # Precision of CP
    R_p=X["R_p"] # Recall of CP
    F05_p=X["F0.5_p"] # F0.5 score of CP
    Fsp_p=X["Fsp_p"] # Fsp score of CP

    # intrinsic metrics of Co with respect to the graph on which it was constructed
    # A node i is adherent to a module m if i ∉ m and i is a neighbor of all the nodes of m
    omega_o=X["omega_o"] # // sum(|adherent(m)|) for m in CO
    TP_o=X["TP_o"] # True positive of CO
    TN_o=X["TN_o"] # True negative of CO
    FP_o=X["FP_o"] # False positive of CO
    FN_o=X["FN_o"] # False negative of CO
    P_o=X["P_o"] # Precision of CO
    R_o=X["R_o"] # Recall of CO
    F05_o=X["F0.5_o"] # F0.5 score of CO
    Fsp_o=X["Fsp_o"] # Fsp score of CO
    # -------------------------------------------------------------
    print(commandline)
    print("|C|=%s, Time.Cp=%s, Time.Co=%s\n"%(str(len(Cp)), str(secondsP), str(secondsO)));
    # -------------------------------------------------------------
    if gnbv_call<=200:
        for i in range(len(Cp)):
            print("m=%s, Ext=%s"%(Cp[i], Ext[i]));
        print("")
    # -------------------------------------------------------------
    print("    |TP_p|=%s, |TN_p|=%s, |FP_p|=%s, |FN_p|=%s"%(TP_p, TN_p, FP_p, FN_p))
    print("    P_p=%s, R_p=%s, F05_p=%s, Fsp_p=%s"%(P_p, R_p, F05_p, Fsp_p))
    print("    omega_p=%s\n"%(omega_p))

    print("    |TP_o|=%s, |TN_o|=%s, |FP_o|=%s, |FN_o|=%s"%(TP_o, TN_o, FP_o, FN_o))
    print("    P_o=%s, R_o=%s, F05_o=%s, Fsp_o=%s"%(P_o, R_o, F05_o, Fsp_o))
    print("    omega_o=%s"%(omega_o))
    # -------------------------------------------------------------
    
    print("========================================================================")
    print("Intrinsic Metrics: Cp")
    print("========================================================================")
    sigma=0.5
    EXIT, ERROR, X = nPnB.TPTNFPFN_Intrinsic(InfileGraph=InfileGraph, InfileClust=OutfileClust, PO="P")
    if (EXIT!=0):
        print(ERROR)
    else:
        TP=X["TP"]; TN=X["TN"]; FP=X["FP"]; FN=X["FN"];
        P, R, F = nPnB.PRF(sigma, TP, TN, FP, FN)
        print("    TP=%s, TN=%s, FP=%s, FN=%s"%(TP, TN, FP, FN))
        print("    Precision(G,Cp)=%s"%(str(P)))
        print("    Recall(G,Cp)=%s"%(str(R)))
        print("    Fscore(G,Cp,sigma=%s)=%s"%(sigma, str(F)))

    print("========================================================================")
    print("Intrinsic Metrics: CO")
    print("========================================================================")
    sigma=0.5
    EXIT, ERROR, X = nPnB.TPTNFPFN_Intrinsic(InfileGraph=InfileGraph, InfileClust=OutfileClust, PO="O")
    if (EXIT!=0):
        print(ERROR)
    else:
        TP=X["TP"]; TN=X["TN"]; FP=X["FP"]; FN=X["FN"];
        P, R, F = nPnB.PRF(sigma, TP, TN, FP, FN)
        print("    TP=%s, TN=%s, FP=%s, FN=%s"%(TP, TN, FP, FN))
        print("    Precision(G,CO)=%s"%(str(P)))
        print("    Recall(G,CO)=%s"%(str(R)))
        print("    Fscore(G,CO,sigma=%s)=%s"%(sigma, str(F)))

    print("========================================================================")
    print("Extrinsic Metrics: Cp/Gold=Cp")
    print("========================================================================")
    sigma=0.5
    EXIT, ERROR, X = nPnB.TPTNFPFN_Extrinsic(InfileGold=OutfileClust, POgold="P",
                                             InfileClust=OutfileClust, POclust="P",
                                             n_vertices=gnbv_call)
    if (EXIT!=0):
        print(ERROR)
    else:
        TP=X["TP"]; TN=X["TN"]; FP=X["FP"]; FN=X["FN"];
        P, R, F = nPnB.PRF(sigma, TP, TN, FP, FN)
        print("    TP=%s, TN=%s, FP=%s, FN=%s"%(TP, TN, FP, FN))
        print("    Precision(Gold=Cp,Cp)=%s"%(str(P)))
        print("    Recall(Gold=Cp,Cp)=%s"%(str(R)))
        print("    Fscore(Gold=Cp,Cp,sigma=%s)=%s"%(sigma, str(F)))

    print("========================================================================")
    print("Extrinsic Metrics: Cp/Gold=Co")
    print("========================================================================")
    sigma=0.5
    EXIT, ERROR, X = nPnB.TPTNFPFN_Extrinsic(InfileGold=OutfileClust, POgold="O",
                                             InfileClust=OutfileClust, POclust="P",
                                             n_vertices=gnbv_call)
    if (EXIT!=0):
        print(ERROR)
    else:
        TP=X["TP"]; TN=X["TN"]; FP=X["FP"]; FN=X["FN"];
        P, R, F = nPnB.PRF(sigma, TP, TN, FP, FN)
        print("    TP=%s, TN=%s, FP=%s, FN=%s"%(TP, TN, FP, FN))
        print("    Precision(Gold=Co,Cp)=%s"%(str(P)))
        print("    Recall(Gold=Co,Cp)=%s"%(str(R)))
        print("    Fscore(Gold=Co,Cp,sigma=%s)=%s"%(sigma, str(F)))

```
