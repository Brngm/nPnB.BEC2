```python
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
# IMPORTS 
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@

import os
import sys
import subprocess

# -----------------------------------------------------------------------------
beep = lambda x: os.system("echo -n '\a';sleep 0.1;" * x)
here=os.path.abspath(os.getcwd())+"/"
print("here ='%s'"%here)
# -----------------------------------------------------------------------------

# =============================================================================
sys.path.append(here+"Graph-Visualization/")
import libraryVisualGraph as Vis

sys.path.append(here+"nPnB/")
import nPnBinterface as nPnB
import GraphFormatConverter as GFC
# =============================================================================

sys.path
devnull = open(os.devnull, 'w')
sys.stderr=devnull
_unused=beep(1) # Everything is well loaded
```

```python
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
# FOLDERS
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
InfolderGraph = "./data/" # The folder in which the networks are stored
OutfolderHTML = "./Graph-Visualization/VISU-HTML/" # The folder in which the html pages will be stored 
```

```python
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
# INPUTS
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
# The name of the input graph
nameGraph="miserables.nPnB.txt" # GraphExample.txt     miserables.nPnB.txt     zachary-karate-club.nPnB.txt     GoodGraph.nPnB.txt

# The name of the output html page
nameHtml="VisCGDSD.html" # "VisCGDSD" for "Visualization of the Clustered Graph at Different Scales of Description"

# The smaller Epsilon, the better the quality of the output clustering, but the slower the computing.
Epsilon=0.01; # in [0,1] (default 0.01)

# Your own description scales: a list of scales in [O,1]
My_scales=[0.5] 

# Intrinsic Scales
IS=True 
step=0.001; # in [0,1] The steps for the search for intrinsic scales (default: 0.001)
# Two clusterings will be considered equivalent if they are equal ("equal") or have the same number of modules ("len" is faster).
eqon="equal"; # "equal" or "len" (default: "equal")

# Your favorite browser
browser="firefox"; # 'firefox' or 'google-chrome' or ...
```

```python
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
# ACTION
# @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
InfileGraph = InfolderGraph + nameGraph # The file in which the network is stored
OutfileHtml = OutfolderHTML + nameGraph + "." + nameHtml # The file in which html page will be stored
# If OutfileHtml already exists, it will be overwritten.

# -----------------------------------------------------------------------------
# Computes clusterings at different scales of description
C=[] # The list of clusterings at different scales of descriptions

# intrinsic scales
if IS:
    C = nPnB.clust_intrinsic(InfileGraph, Epsilon=Epsilon, step=step, eqon=eqon)
    print("%s intrinsic scales"%(len(C)))

# My_scales
SI = [c["input_sp"] for c in C]
C = C + nPnB.clust_scale(InfileGraph, [s for s in My_scales if not s in SI])
# -----------------------------------------------------------------------------

# -----------------------------------------------------------------------------
# Builds the html page
Vis.make3DHTML(GFC.GRAPH_nPnB_2_json(InfileGraph), C, OutfileHtml)
# -----------------------------------------------------------------------------

# -----------------------------------------------------------------------------
# Launch the Visualization of the Clustered Graph at Different Scales of Description
process=subprocess.Popen([browser, f'file://{os.path.join(here, OutfileHtml)}']) 
# -----------------------------------------------------------------------------
```
